## 3.	Tones per Gallon per distance travelled matrix for each truck and all trucks.

```python
# it is for each truck 
df_truck['tones_per_gallon_per_distance'] = df_truck['PAYLD_WT'] / df_truck['FUEL_USED_IN_CYC'] / df_truck['LD_TRAV_DSTNC']
```

```python
# for all trucks or overall
overall_tones_per_gallon_per_distance =  df_truck['PAYLD_WT'].sum() / df_truck['FUEL_USED_IN_CYC'].sum() / df_truck['LD_TRAV_DSTNC'].sum()
```

```python
overall_tones_per_gallon_per_distance
```

```python
# for each truck

def tones_per_gallon_per_distance_of_truck(name_truck):

    pay_load = df_truck.loc[df_truck['MACH_SER_NO'] == name_truck, 'PAYLD_WT'].sum()
    fuel = df_truck.loc[df_truck['MACH_SER_NO'] == name_truck, 'FUEL_USED_IN_CYC'].sum()
    distance = df_truck.loc[df_truck['MACH_SER_NO'] == name_truck, 'LD_TRAV_DSTNC'].sum()

    return (pay_load/fuel/distance)
```

## 4.	Which truck is running most efficient and which truck is running most inefficient as per as the matrix in item number 3?

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.bar(df_truck['MACH_SER_NO'], df_truck['tones_per_gallon_per_distance'])
plt.xlabel('Truck')
plt.ylabel('Tones per Gallon per distance')
plt.title('Tones per Gallon per distance traveled for each truck')
plt.xticks(rotation=45)
plt.show()
```

#### According to this we can clearly see that Truck 16 is most efficient while truck 10 is inefficient

```python


# Sorting the DataFrame by empty travel distance in descending order
sorted_df = df_truck.sort_values(by='EMTY_TRAV_DSTNC', ascending=False)

# Plotting the empty travel distance for each truck
plt.figure(figsize=(10, 6))
plt.barh(sorted_df['MACH_SER_NO'], sorted_df['EMTY_TRAV_DSTNC'])
plt.xlabel('Empty Travel Distance (km)')
plt.ylabel('Truck')
plt.title('Empty Travel Distance for Each Truck')
plt.show()
```

## 6.	Show visually which truck has Max and Min Gear Shifts per Mile

```python
b.info()
```

```python
# Selecting the desired columns from df_B
selected_columns = b[['Total_distance', 'Miles', 'Gear_shift_Miles']]

# Merging the selected columns into df_A
merged_df = pd.concat([df_truck, selected_columns], axis=1)
```
